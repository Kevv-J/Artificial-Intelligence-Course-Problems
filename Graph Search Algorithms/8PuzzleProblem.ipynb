```python
def getInvCount(arr):
    invCount = 0
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            if int(arr[j]) and int(arr[i]) and int(arr[i]) > int(arr[j]):
                invCount += 1
    
    return invCount
            
def isSolvable(initial,goal):
    invCount1 = getInvCount(initial)
    invCount2 = getInvCount(goal)
    return (invCount1%2 == invCount2%2) 

# print(isSolvable("123456780"))

```

```python
def indexOfGoal(goal):
    a = [0 for x in range(len(goal)-1)]
    for i,x in enumerate(goal):
        if int(x):
            ind = goal.index(x)
            a[int(x)-1]=(ind//3,ind%3)
    return a

# indexOfGoal("876543210")
```

```python
def manhattanDist(arr,goal):
    dist = 0
    a=[]
    b=indexOfGoal(goal)
    a.append([x for x in arr[0:3]])
    a.append([x for x in arr[3:6]])
    a.append([x for x in arr[6:9]])
    for i in range(len(a)):
        for j in range(len(a[0])):
            if int(a[i][j]):
                dist += abs(i-b[int(a[i][j])-1][0])
                dist += abs(j-b[int(a[i][j])-1][1])
    return dist

# print(manhattanDist("654321870","876543210"))
```

```python
def printPuzzle(arr):
    print('\n')
    print([x for x in arr[0:3]])
    print([x for x in arr[3:6]])
    print([x for x in arr[6:9]])
    
def eSwap(a,i,j):
    a[i],a[j] = a[j],a[i]
```

```python
def stateSpaceExpansion(arr):
    ind = arr.index('0')
    s = [c for c in arr]
#     print(ind,s)

#     printPuzzle(arr)
    
    if ind == 0:
        l = []
        a = s.copy()
        eSwap(a,0,1)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,0,3)
        l.append("".join(a))
        return l
    if ind == 1:
        l = []
        a = s.copy()
        eSwap(a,1,0)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,1,2)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,1,4)
        l.append("".join(a))
        return l
    if ind == 2:
        l = []
        a = s.copy()
        eSwap(a,2,1)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,2,5)
        l.append("".join(a))
        return l
    if ind == 3:
        l = []
        a = s.copy()
        eSwap(a,3,0)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,3,4)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,3,6)
        l.append("".join(a))
        return l
    if ind == 4:
        l = []
        a = s.copy()
        eSwap(a,4,1)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,4,3)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,4,5)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,4,7)
        l.append("".join(a))
        return l
    if ind == 5:
        l = []
        a = s.copy()
        eSwap(a,5,2)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,5,4)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,5,8)
        l.append("".join(a))
        return l
    if ind == 6:
        l = []
        a = s.copy()
        eSwap(a,6,3)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,6,7)
        l.append("".join(a))
        return l
    if ind == 7:
        l = []
        a = s.copy()
        eSwap(a,7,4)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,7,6)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,7,8)
        l.append("".join(a))
        return l
    if ind == 8:
        l = []
        a = s.copy()
        eSwap(a,8,5)
        l.append("".join(a))
        a = s.copy()
        eSwap(a,8,7)
        l.append("".join(a))
        return l

# stateSpaceExpansion("876540213")
```

```python
def route(start,goal,explored):
    node = [x for x in explored if x[0] == goal]
    node = node[0]
    cost = node[3]+node[1]
    
    path = []
    while node[0] != start:
        path.insert(0,node[0])
        node = [x for x in explored if x[0] == node[4]]
        node = node[0]
    path.insert(0,start)
    return path,cost
        
```

```python
def Search(start,goal,algo): # algo Can be ["DFS","BFS","Greedy","AStar"]
    frontier = [(start,0,manhattanDist(start,goal),0,None)]
    explored = set()
    while True:
        if algo == "AStar":
            frontier = sorted(frontier, key=lambda x: x[1]+x[2]+x[3])
        if algo == "Greedy":
            frontier = sorted(frontier, key=lambda x: x[2])
        if not frontier:
            return ("Failure",-1)
        if goal in [x[0] for x in explored]:
            return route(start,goal,explored)
        if algo == "DFS":
            node = frontier.pop()
        else :
            node = frontier.pop(0)
        if node[0] not in [x[0] for x in explored]:
            explored.add(node)
        neighbours = stateSpaceExpansion(node[0])
        for neighbour in neighbours:
            if neighbour not in [x[0] for x in explored]:
                new_node = (neighbour,1,manhattanDist(neighbour,goal),node[1]+node[3],node[0])
                frontier.append(new_node)
```
